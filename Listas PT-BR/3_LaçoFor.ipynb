```python
'''15.Contagem de Números Múltiplos de 3 e 5'''
#entrada usuário
numUsuario = int(input("Digite um número: "))
#armazenamento dos valores
mult3 = 0
mult5 = 0
#laço for
for i in range(1, numUsuario + 1):
    if i % 3 == 0:
        mult3 += 1
        print(i, "é múltiplo de 3")

    if i % 5 == 0:
        mult5 += 1
        print(i, "é múltiplo de 5")
#print
print("\nQuantidade de múltiplos de 3:", mult3)
print("Quantidade de múltiplos de 5:", mult5)


```

```python
'''16.Peça ao usuário 10 números e exiba quantos são positivos, negativos ou zero.'''
#armazenamento dos valores
positivos = 0
negativos = 0
zero = 0
#laço for
for i in range(10):
    numerosUsuario = int(input("Digite um numero: "))

    if numerosUsuario > 0:
        positivos += 1
    elif numerosUsuario < 0:
        negativos += 1
    else:
        zero += 1
#print
print(positivos)
print(negativos)
print(zero)

```
