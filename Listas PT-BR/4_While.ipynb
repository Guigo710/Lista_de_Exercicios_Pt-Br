```python
""" 1.Soma de Números: Crie um programa que solicite ao usuário para inserir números e some esses números até que o usuário insira zero.
Quando zero for inserido, o programa deve imprimir a soma total."""
#contador
soma = 0
#entrada usuário
numero = float(input('Digite um número (0 para sair): '))
#estrutura de decisão
while numero != 0:
  soma += numero
  numero = float(input("Digite outro número: "))
#print do resultado
print("Soma total:",soma)
```

```python
"""2.Encontrar o Máximo: Crie um programa que peça ao usuário para inserir números e encontre o maior número inserido.
 O programa deve continuar pedindo números até que o usuário digite "sair"."""
#contador
maior = None

while True:
    usuario = input('Digite um número (digite "sair" para encerrar): ')
#condição pra sair
    if usuario.lower() == 'sair':
        break
#estrutura de decisão considerando o erro do usuário ao digitar algo além de número.
    try:
  #transforma o que o usuário digitou (texto) em número decimal.
      numero = float(usuario)
#lógica
      if maior is None or numero > maior:
        maior = numero
#print do erro ao digitar letra em vez de número
    except ValueError:
        print('Entrada inválida! Digite um número ou "sair".')
#print do resultado
print(f'O maior número digitado foi: {maior}')

```

```python
"""3.Quantidade de valores: Conte quantos valores positivos, negativos e zeros foram digitados.""" #Adcionei os valores também
#contadores
positivos = []
negativos = []
zeros = []

while True:
  numeros = input("Digite um valor (digite 'sair' para encerrar): ")
#condição pra sair
  if numeros.lower() == 'sair':
    break
#utilizão do try e except para erro de digitação
  try:
    num = float(numeros)
#estrutura de decisão
    if num > 0:
      positivos.append(num)
    elif num < 0:
      negativos.append(num)
    else:
      zeros.append(num)
  except ValueError:
      print('Opção inválida')
#print com quantidade e os valores
print(f'Quantidade de positivos: {len(positivos)} | {positivos}')
print(f'Quantidade de negativos: {len(negativos)} | {negativos}')
print(f'Quantidade de zeros: {len(zeros)} | {zeros}')

```

```python
"""4.Conversão de Temperatura: Faça um programa que converta uma temperatura de Celsius para Fahrenheit.
Continue pedindo ao usuário para inserir uma nova temperatura em Celsius até que ele digite "sair"."""
#Lógica
while True:
  temperatura = input("Digite a temperatura em Celsius(digite 'sair' para encerrar): ")
  if temperatura.lower()== 'sair':
    break
#try e except para caso o usuário digite algo errado
  try:
    temp = float(temperatura)
    Fahrenheit = temp * 1.8 + 32
    print(f'{temperatura}°C em Fahrenheit é: {Fahrenheit}')
  except ValueError:
    print('Opção inválida')



```

```python
"""5.Simulador de Caixa Eletrônico: Crie um programa que simule um caixa eletrônico,
que continue pedindo ao usuário para inserir um valor de saque até que o saldo da conta seja zero."""
#armazenamento de valor
saldo = 800
#Condição para parar
while True:
  saque = input("Digite um valor de saque: ")
  if saldo <= 0:
    print("Saldo esgotado.")
    break
#Lógica e estrutura de decisão
  try:
      valor = float(saque)

      if valor <= 0:
          print("Digite um valor válido.")
      elif valor > saldo:
          print("Saldo insuficiente.")
      else:
          saldo -= valor
          print(f"Saque realizado. Saldo atual: R$ {saldo}")

  except ValueError:
        print("Entrada inválida. Digite um número.")

```

```python
"""6.Média de Notas: Escreva um programa que continue pedindo ao usuário para inserir notas (0 a 10) e calcule a média dessas notas.
O programa deve parar quando o usuário digitar uma nota negativa."""
#lista para guardar as notas
notas = []
#condição verdadeira
while True:
    entrada = input('Digite sua nota (nota negativa encerra): ')
#transformando a nota de string para float
    try:
        nota = float(entrada)
#condição para o break
        if nota < 0:
            break
#deixando apenas notas de 0 a 10
        if nota > 10:
            print('Nota inválida. Digite um valor entre 0 e 10.')
 #adcionando as notas na lista
        else:
            notas.append(nota)
#print do erro
    except ValueError:
        print('Entrada inválida. Digite um número.')
#lógica da soma e da quantidade de notas e erro do usuario
if len(notas) > 0:
    media = sum(notas) / len(notas)
    print(f'Notas digitadas: {notas}')
    print(f'Média das notas: {media}')
else:
    print('Nenhuma nota válida foi digitada.')

```

```python
"""7.Conversor de Moeda: Crie um programa que converta uma quantia em dólares para euros.
Continue pedindo ao usuário quantias em dólares para converter até que ele insira "0"."""
#cotação do dia 1 dólar = 0.84 euro
while True:
  entrada = input("Digite o valor que você deseja converter(inserir 0 encerra): ")
  if entrada == '0':
    break
  try:
      valor = float(entrada)
      conversao = valor * 0.84
      print(f'${entrada} convertido para Euro: €{conversao}')
  except ValueError:
      print('Opção inválida, digite um valor')
```

```python
"""8.Controle de Estoque: Faça um controle de estoque com menu de entrada, saída e exibição."""
#quantidade no estoque
estoque = 5
#logica e estrutura de decisão
while True:
  opcao = input("Digite qual a opção que deseja:\n 1.Entrada \n 2.Saída \n 3.Consulta\n 4.Sair do programa\n")
  if opcao == '1':
    qntd_entrada = float(input('Digite a quantidade para entrar no estoque: \n'))
    estoque += qntd_entrada
    print(f'A nova quantidade do estoque é: {estoque}')
  elif opcao == '2':
    qntd_saida = float(input('Digite a quantidade para sair do estoque:  \n'))
    if qntd_saida > estoque:
      print('Quantidade indisponivel')
    else:
        estoque -= qntd_saida
    print(f'A nova quantidade do estoque é: {estoque}')
  elif opcao == '3':
    print(f'A quantidade disponivel é de: {estoque}')
  else:
    break

```

```python
"""9.Simulação de Dados de Sensor: Crie um programa que simule a leitura de dados de um sensor e
continue capturando dados até que um valor fora do intervalo de operação seja detectado (por exemplo, fora de 0 a 100)."""
#importando um "simulador de dados"
import random
#lógica da estrutura
while True:
  leitura = random.randint(-20,120)
  print(f'Leitura do sensor:{leitura}')
#declarando o range que vai encerrar a leitura
  if leitura in range(0,101):
    print(f'Valor fora do intervalo de operação. Encerrando leitura')
    break
```

```python
"""10.Controle de Votação: Faça um programa que permita cadastrar votos para 3 candidatos. Exibe contagem ao final quando for digitado "fim"."""
#contador
candidato1= 0
candidato2= 0
candidato3= 0
#lógica e estrutura de decisão
while True:
  votos = input("Digite numericamente o candidato que deseja votar:\n 1.candidato 1 \n 2.candidato 2\n 3.candidato 3\n")
  if votos =='1':
    candidato1 += 1
    print(f'Voto registrado para o Candidato 1')
  elif votos == '2':
    candidato2 += 1
    print(f'Voto registrado para o Candidato 2')
  elif votos == '3':
    candidato3 += 1
    print(f'Voto registrado para o Candidato 3')
  #encerrar o loop
  elif votos.lower() == 'fim':
    print(f'Encerrando contagem')
    break
  #erro do usuário
  else:
    print(f'Opção inválida. tente novamente')
print(f'A votação se encerrou e ficou assim: \nCandidato 1: {candidato1} \n Candidato 2: {candidato2}\n Candidato 3: {candidato3}')

```
