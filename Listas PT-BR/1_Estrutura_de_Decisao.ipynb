```python
'''4.Crie uma calculadora simples que permita ao usuário escolher uma operação
(adição, subtração, multiplicação ou divisão) e dois números, e então exiba o resultado.'''

#entrada usuário
Valor_a = float(input("Digite um número: "))
Escolha_operação= str(input("Escolha seu operação:'+','-','*','/':"))
Valor_b= float(input("Digite outro número: "))
#função
if Escolha_operação == "+":
    soma = Valor_a + Valor_b
    print(f'A soma é:{soma}.')
elif Escolha_operação == '-':
    subtração = Valor_a - Valor_b
    print(f'A subtração é: {subtração}')
elif Escolha_operação =='*':
    multiplacação = Valor_a * Valor_b
    print(f'A multiplicação é: {multiplacação}.')
elif Escolha_operação =='/':
    divisão = Valor_a / Valor_b
    print(f'A divisão é: {divisão}.')
else:
    print('Por favor escolha a operação correta.')

```

```python
'''8.Peça ao usuário seu peso e altura, calcule o Índice de Massa Corporal (IMC) e classifique-o como
"Abaixo do peso",
"Peso normal",
"Sobrepeso"
"Obesidade".'''
#entrada usuário
Peso = float(input("Digite seu peso: "))
Altura = float(input("Digite sua altura: "))
#Função
IMC = Peso / (Altura**2)
#Chamando a função
if IMC < 18.5 :
  print(f'Seu IMC é de: {IMC: .2f} magreza')
elif IMC >=18.5 and IMC < 24.9:
  print(f'Seu IMC é de: {IMC: .2f} Normal')
elif IMC >=25.0 and IMC < 29.9:
  print(f'Seu IMC é de: {IMC: .2f} Sobrepeso')
elif IMC >=30.0 and IMC < 39.9:
  print(f'Seu IMC é de: {IMC: .2f} Obesidade')
else:
  print(f'Seu IMC é de: {IMC: .2f} obesidade grave')

```

```python
'''12 - Sistema de Hotel – Verificação de Diárias Um sistema de hotel deve calcular o valor total da hospedagem.
O usuário informa o número de diárias. Cada diária custa R$ 120. Se o número de diárias for maior que 7, o
 sistema aplica um desconto de 10% no total. Desenvolva esse algoritmo com uso de operadores e estrutura de decisão.'''
#entrada usuário
num_diarias = float(input("Digite a quantidade de diárias: "));
#Função
diaria = 120 * num_diarias

if num_diarias >= 7:
    valor_desconto = diaria - (diaria * 0.10)
    print(f'O valor com o desconto é de: R${valor_desconto:.2f}.')
else:
  print(f'O valor da diária sem o desconto de 10% é: {diaria:.2f}')

```

```python
''' 13; - Sistema de Restaurante – Pedido com Adicional Um sistema de restaurante oferece dois pratos:
 • Prato A: R$ 20 • Prato B: R$ 25 O cliente pode ainda adicionar uma bebida por R$ 5.
 Desenvolva um algoritmo que pergunte qual prato foi escolhido (A ou B), se deseja bebida (S ou N) e calcule o valor total do pedido.'''

#entrada usuario
pedido = input('Qual prato voce deseja\n•Prato A: R$ 20 \n•Prato B: R$ 25\n').lower()
bebida = input('Deseja adicionar uma bebida por R$ 5? (S ou N)\n').lower()
#estrutura de decisão
total = 0
#pedido
if pedido == 'a':
  total = 20
elif pedido == 'b':
  total = 25
else:
  print(f'Opção inválida.')
#bebida
if bebida =='s':
  total = total + 5
elif bebida == 'n':
  total = total
else:
  print(f'Opção inválida.')
#chamando o print com o valor total
print(f'O valor total do seu pedido é: R$ {total}')
```
