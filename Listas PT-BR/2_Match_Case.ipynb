```python
"""1. Conversor de Moedas

Implemente um conversor de moedas que permita ao usuário escolher entre
Dólar (USD), Euro (EUR) e Libra (GBP) e converter um valor informado para Reais (BRL)."""

#entrada usuario
moeda = int(input("Escolha a conversão para a moeda desejada:\n 1.Dólar  \n 2.Euro \n 3.Libra \n"))
valor = float(input("Digite o valor desejado: "))

#formula
dolar = 5.30
euro = 6.27
libra = 7.23
match moeda:
  case 1:
    print(f'A conversão de real para Dólar é de:{dolar * valor:.2f}')
  case 2:
    print(f'A conversão de real para Euro é de:{euro * valor:.2f}')
  case 3:
    print(f'A conversão de real para Libra é de:{libra * valor:.2f}')
```

```python
"""2 . Cálculo de Área de Figuras Geométricas

Desenvolva um programa que permita ao usuário escolher entre calcular
a área de um quadrado, retângulo ou triângulo e insira os valores necessários para o cálculo."""

#entrada usuario
forma_geometrica = int(input("Escolha a forma geometrica desejada:\n 1.quadrado  \n 2.retângulo \n 3.triângulo \n"))

#formulas
"""area_quadrado = lado * lado
area_triangulo = base * altura/2
area_retangulo = base * altura """

match forma_geometrica:
  case 1:
    lado = float(input("Digite o valor do seu lado:"))
    print(f"A área do seu quadrado é de: {lado * lado:.2f}")
  case 2:
    base = float(input("Digite o valor da sua base:"))
    altura = float(input("Digite o valor da altura:"))
    print(f"A área do seu retângulo é de: {base * altura:.2f}")
  case 3:
    base = float(input("Digite o valor da sua base:"))
    altura = float(input("Digite o valor da altura:"))
    print(f"A área do seu triângulo é de: {base * altura/2:.2f}")
```

```python
"""5 . Tradutor de Cores

O usuário informa uma cor em português (vermelho, azul, verde, amarelo), e o programa exibe sua tradução para inglês."""


#entrada usuario
cores = int(input("Escolha a sua cor:\n 1.Vermelho  \n 2.Azul \n 3.Verde \n 4.Amarelo \n"))
#estrutura de decisão
match cores:
  case 1:
    print(f'A cor Vermelho em inglês é Red.')
  case 2:
    print(f'A cor Azul em inglês é Blue.')
  case 3:
    print(f'A cor Verde em inglês é Green.')
  case 4:
    print(f'A cor Amarelo em inglês é Yellow.')
```

```python
"""6. Simulação de Atendimento Telefônico

O usuário seleciona uma opção de atendimento telefônico:
1 - Suporte Técnico
2 - Financeiro
3 - Cancelamento
4 - Falar com um atendente"""

#Entrada usuário:
opcao = int(input("Escolha a opção para o atendimento telefônico:\n 1.Suporte Técnico \n 2.Financeiro \n 3.Cancelamento \n 4.Falar com um atendente \n"))
#estrutura de decisão:
match opcao:
  case 1:
    print(f'Em instantes você será redirecionado ao Suporte Técnico.')
  case 2:
    print(f'Em instantes você será redirecionado ao setor Finaceiro.')
  case 3:
    print(f'Em instantes você será redirecionado ao setor responsável pelo Cancelamento.')
  case 4:
    print(f'Em instantes você será redirecionado para um de nossos atendentes.')
```

```python
""" 7. Calculadora Simples

O usuário escolhe uma operação matemática (+, -, *, /) e insere dois números. O programa exibe o resultado. """
# Entrada usuário:
operacao = int(input("Escolha uma operação que você deseja realizar:\n 1:+ \n 2:- \n 3:* \n 4:/ \n"))
num1 = float(input("Insira o primeiro número: "))
num2 = float(input("Insira o segundo número: "))
#estrutura de decisão
match operacao:
  case 1:
    print(f'A adição do {num1} + {num2} é: {num1 + num2}')
  case 2:
    print(f'A subtração do {num1} - {num2} é: {num1 - num2}')
  case 3:
    print(f'A multiplicação do {num1} * {num2} é: {num1 * num2}')
  case 4:
    print(f'A divisão do {num1} / {num2} é: {num1 / num2:.2f}')
```

```python
'''8. Sistema de Reserva de Passagens

O usuário escolhe um destino (São Paulo, Rio de Janeiro, Curitiba, Salvador) e recebe o valor da passagem.'''

#Entrada do usuário
destino = int(input('Escolha um destino:\n 1.São Paulo \n 2.Rio de Janeiro \n 3.Curitiba \n 4.Salvador \n '))
#valores fixados
sao_paulo = 250.00
rio_de_janeiro = 300.00
curitiba = 150.00
salvador = 650.00
#estrutura de decisao:
match destino:
  case 1:
    print(f'O valor da passagem para São Paulo é de R${sao_paulo}')
  case 2:
    print(f'O valor da passagem para Rio de Janeiro é de R${rio_de_janeiro}')
  case 3:
    print(f'O valor da passagem para Curitiba é de R${curitiba}')
  case 4:
    print(f'O valor da passagem para Salvador é de R${salvador}')


```
